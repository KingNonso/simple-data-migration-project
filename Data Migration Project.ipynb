```python
# Filter by only where Verification successful
filtered_df = payments[payments['note'] == 'Verification successful']
# Create a list of indices or row labels that holds important payment details
payments_df = filtered_df[['payment_id', 'user_id', 'amount', 'purpose', 'ntype_id', 'method', 'ImageUrl', 'note', 'date']]
payments_df.tail(10)
```

```python
# Get all unique categories from the 'purpose' column
unique_categories = payments_df['purpose'].unique()
unique_notes = payments_df['note'].unique()
print(unique_categories)
print(unique_notes)
```

```python
# Group payments by user_id
grouped_payments = payments_df.groupby('user_id')

grouped_payments.head()
```

```python
# Display groups
for user_id, group in grouped_payments:
    print(f"User ID: {user_id}")
    print(group)
    print("\n")
```

```python
# Define the start and end dates for each period
periods = [
    (datetime(2020, 7, 1), datetime(2021, 6, 30)),
    (datetime(2021, 7, 1), datetime(2022, 6, 30)),
    (datetime(2022, 7, 1), datetime(2023, 6, 30)),
    (datetime(2023, 7, 1), datetime(2024, 6, 30))
]

# Initialize payment_df with unique user IDs
transactions = pd.DataFrame({'user_id': grouped_payments.groups.keys()})

print(transactions)

```

```python
# Iterate over periods and update payment_df based on conditions
for start_date, end_date in periods:
    # Apply aggregation function to grouped DataFrame
    period_payments = grouped_payments.apply(lambda x: x[(x['date'] >= start_date.date()) & (x['date'] <= end_date.date())]).reset_index(drop=True)
    
    # Convert 'date' column to datetime
    period_payments['date'] = pd.to_datetime(period_payments['date'])
    
    # Create another column for the full date and clone date into it
    period_payments['full_date'] = period_payments['date'].dt.strftime('%Y-%m-%d')

    # Group by 'user_id' and sum for each year
    sum_per_year = period_payments.groupby(['user_id', period_payments['date'].dt.year]).agg(
    {'amount': 'sum', 'full_date': 'first', 'purpose': 'first'}).reset_index()
    
    # date columns being renamed here to allow for data import via the API
    periodic_date = f"two{start_date.strftime('%y')}{end_date.strftime('%y')}"
    print('date', periodic_date)
    # Iterate over unique user IDs in transactions
    for user_id in transactions['user_id']:
        # Check if the user ID exists in sum_per_year
        if user_id in sum_per_year['user_id'].values:
            # Get the corresponding row from sum_per_year
            user_row = sum_per_year[sum_per_year['user_id'] == user_id]
            # Assign the sum of amounts for the current period to the corresponding column in transactions
            transactions.loc[transactions['user_id'] == user_id, str(periodic_date) ] = user_row['amount'].values[0]
            transactions.loc[transactions['user_id'] == user_id, f'{periodic_date}date' ] = user_row['full_date'].values[0]
        else:
            # If the user ID doesn't exist in sum_per_year, assign 0 to the corresponding column in transactions
            transactions.loc[transactions['user_id'] == user_id, f'{periodic_date}'] = 0
            transactions.loc[transactions['user_id'] == user_id, f'{periodic_date}date' ] = ''
        

print(transactions)
```

## Make GraphQL request
transform the output of the pandas DataFrame into a GraphQL request payload efficiently using Python string formatting

```python
import requests
import json
```

```python
# Loop through DataFrame rows and create GraphQL request payload for each row
counter = 1
for _, row in data.iterrows():
    # Create input variables dictionary
    input_variables = {
        'orgId': "2",
        'userId': str(row['user_id']),
        'email': row['email'],
        'orgNumber': row['org_number'],
        'lastName': row['last_name'],
        'firstName': row['first_name'],
        'middleName': row['middle_name'],
        'phone': row['phone'],
        'nseNumber': row['nse_number'],
        'corenNumber': row['coren_number'],
        'organisation': row['organisation'],
        'gender': row['gender'],
        'nseCategory': row['nse_category'],
        
        'yearsOfEngExperience': row['years_of_eng_experience'],
        'yearsOfPgExperience': row['years_of_pg_experience'],
        
        'dob': row['dob'],
        'levelOfResponsibility': row['level_of_responsibility'],
        'bio': row['bio'],
        'emplomentStatus': row['emploment_status'],
        'membershipType': row['membership_type'],
        'usersStatusType': row['users_status_type'],
        'usersPracticeType': row['users_practice_type'],
        'usersSpecialtyType': row['users_specialty_type'],
        'usersSpecialtyId': row['users_specialty_id'],
        'membershipTypeId': row['membership_type_id'],
        'otherMembershipNumbers': row['other_membership_numbers'],
        'two2021': row['two2021'],
        'two2021date': row['two2021date'],
        'two2122': row['two2122'],
        'two2122date': row['two2122date'],
        'two2223': row['two2223'],
        'two2223date': row['two2223date'],
        'two2324': row['two2324'],
        'two2324date': row['two2324date'],
        
    }
    
    # Construct the full GraphQL request payload
    formatted_variables = {}
    for key, value in input_variables.items():
        
        if value is not None and value != '' and not pd.isna(value):
            #print(key, value)
            if isinstance(value, str):
                # If the value is a string, remove newline characters
                formatted_value = value.replace('\n', '')
                formatted_variables[key] = formatted_value
            else:
                # If the value is not a string, convert it to string format
                #pass
                formatted_variables[key] = value
            
        
    variables_str = ", ".join([f"{key}: \"{value}\"" for key, value in formatted_variables.items()])
    graphql_mutation = """
        mutation {{
          migrateData({}) {{
            success
            errors
          }}
        }}
    """.format(variables_str)

    

    #print(graphql_mutation)
    #continue

    
    payload = {'query': graphql_mutation}
    
    # Convert payload to JSON string
    payload_str = json.dumps(payload)
    #print(payload_str)

    # Make the request
    url = "http://localhost:8000/graphql"
    headers = {"Content-Type": "application/json"}
    response = requests.post(url, data=payload_str, headers=headers)

    # Print the response
    print(row['email'], counter)
    print(response.text)
    print('\n')
    counter += 1
    #break
    
```
